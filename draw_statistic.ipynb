```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

hedera_top10_volume_dict = {
    1: '1770000',
    2: '1630000',
    3: '1390000',
    4: '441120',
    5: '285340',
    6: '259930',
    7: '251350',
    8: '235290',
    9: '227100',
    10: '218150'
}

eth_top10_volume_dict = {
    1: '4250000000',
    2: '3890000000',
    3: '3070000000',
    4: '1980000000',
    5: '1570000000',
    6: '1470000000',
    7: '1460000000',
    8: '1290000000',
    9: '1030000000',
    10: '1010000000'
}

hedera_top10_volume_df = pd.DataFrame(hedera_top10_volume_dict.items(), columns=['Rank', 'Volume'])
eth_top10_volume_df = pd.DataFrame(eth_top10_volume_dict.items(), columns=['Rank', 'Volume'])

hedera_top10_volume_df['Volume'] = hedera_top10_volume_df['Volume'].astype(int)
eth_top10_volume_df['Volume'] = eth_top10_volume_df['Volume'].astype(int)

hedera_top10_volume_df['Volume'] = np.log(hedera_top10_volume_df['Volume'])
eth_top10_volume_df['Volume'] = np.log(eth_top10_volume_df['Volume'])

ax = hedera_top10_volume_df.plot(x='Rank', y='Volume', color='#ff7f0e')
eth_top10_volume_df.plot(x='Rank', y='Volume', ax=ax)
# Set y-axis label log(Volume)
ax.set_ylabel('log(Volume)')

# Calculate linear regression for Hedera
hedera_X = hedera_top10_volume_df[['Rank']]
hedera_y = hedera_top10_volume_df['Volume']
hedera_model = LinearRegression().fit(hedera_X, hedera_y)
hedera_top10_volume_df['Linear_Regression'] = hedera_model.predict(hedera_X)

# Calculate linear regression for Ethereum
eth_X = eth_top10_volume_df[['Rank']]
eth_y = eth_top10_volume_df['Volume']
eth_model = LinearRegression().fit(eth_X, eth_y)
eth_top10_volume_df['Linear_Regression'] = eth_model.predict(eth_X)

# Plot linear regression lines
hedera_line, = ax.plot(hedera_top10_volume_df['Rank'], hedera_top10_volume_df['Linear_Regression'], linestyle='--', color=ax.get_lines()[0].get_color())
eth_line, = ax.plot(eth_top10_volume_df['Rank'], eth_top10_volume_df['Linear_Regression'], linestyle='--', color=ax.get_lines()[1].get_color())

# Get slopes
hedera_slope = hedera_model.coef_[0]
eth_slope = eth_model.coef_[0]

# Display slopes on the plot
plt.text(1, hedera_top10_volume_df['Volume'].max(), f'slope = {hedera_slope:.2f}', fontsize=12, color='black')
plt.text(1, eth_top10_volume_df['Volume'].max(), f'slope = {eth_slope:.2f}', fontsize=12, color='black')

ax.legend(['Hedera', 'Ethereum', 'Hedera Linear Regression', 'Ethereum Linear Regression'])
ax.set_title('Hedera and Ethereum Top 10 Volume with Linear Regression')

plt.show()
```
